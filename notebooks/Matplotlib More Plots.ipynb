# Matplotlib: More Plots

There are **many** different types of plots available from Matplotlib. We've already seen the **line** plot. Let's take a look at some other common plot types. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Scatter plot

```python
# make some data
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)
plt.axes([0.025, 0.025, 0.95, 0.95])

# you should change s, c, and alpha to see what they do
plt.scatter(X, Y, s=75, c=T, alpha=0.1)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

# invisible ticks!
plt.xticks(())
plt.yticks(())

plt.show()
```

### Bar plots

In this plot we specify the bar **color** two different ways: (1) using the **hex color code**, which is a 6-digit alpha-numeric code, and (2) using the **color name**, such as "white".  

All Matplotlib plots allow you to assign color using the hex color code or the color name. 

```python
# make some data
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

# you can specify color using the hex code or the name
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

plt.ylim(-1.25, +1.25)

plt.legend(['Duke','NC State'],fontsize=10)

plt.show()
```

### Contour plot

Contour plots are for **3-dimensional** data, such as weather maps. There are two main functions needed to make a contour plot. First, we use `np.meshgrid()` to create the x-y coordinate system. Then, we use `plt.contour()` to plot the data. `plt.contour()` takes **3** main arguments: x, y, and z data. In this example, our z-data is computed from a function, **f**, that we define ourselves.  

```python
# write a function that computes Z given X and Y
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

# make a mesh grid
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

# plot the data. You should change levels, alpha and cmpa to see what they do
plt.contourf(X, Y, f(X, Y), levels=8, alpha=.75, cmap='jet')

# What happens if you delete this line?
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
```

### Imshow: plotting a 2d array

`plt.imshow()` displays matrices as **images**.  In the previous example, we saw that `plt.contour()` plots matrices in **vector** format (smooth lines). `plt.imshow()` is different because it plots data in **raster** format (pixels).

```python
# write a function that computes Z given X and Y
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y ** 2)

# make a mesh grid
n = 10
x = np.linspace(-3, 3, 3 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)

# compute Z
Z = f(X, Y)

# you should change interpolation, cmap, and origin to see what happens
plt.imshow(Z, interpolation='nearest', cmap=plt.get_cmap('terrain'), origin='lower')

# add a colorbar
plt.colorbar(shrink=.92)

# remove ticks
plt.xticks(())
plt.yticks(())

plt.show()
```

### Histogram

**Histograms** are an invaluable way to visualize 1-dimensional data. In the example below, we generate a random sample of normally-distributed data, then plot the data using `plt.hist()`. Since the sample is random, we first call `np.random.seed(1)` to specify the **seed** of the function. This means that every time we run the cell we will get the same result. 

Does it look different if you change the seed number? 

```python
# set the seed
np.random.seed(1)

# take a random sample of data
x = np.random.randn(10000)

# you should change density, bins, facecolor, and alpha to see what they do
plt.hist(x, density=1,bins=50, facecolor='r', alpha=0.5)

# modify the axes and title
plt.xlabel('Values',fontsize=15)
plt.ylabel('Probability',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Standard normal distribution',fontsize=20)

# add some text at coordinates: x=1, y=0.35
plt.text(1,0.35, r'$\mu=0,\ \sigma=1$',fontsize=20)

plt.show()
```

### Pie charts

Everyone loves a **pie chart**!

```python
# make a vector of ones
n = 20
Z = np.ones(n)

# change the second-to-last value to 2
Z[-1] *= 2

# can you figure out how to make it colorful?
plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])

plt.axis('equal')
plt.xticks(())
plt.yticks()

plt.show()
```
