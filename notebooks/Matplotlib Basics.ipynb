Next, load the packages.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Note that we did not import the entire Matplotlib package; rather, we imported a **subset** of the package called "pyplot", which contains all of the basic plotting functionality. 

### Line plot

First, use **Numpy** functions to **generate data** for our plots. You should already be familiar with `np.linspace()`. 

Can you guess what `np.cos()` and `np.sin()` do?

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.cos(x)
y2 = np.sin(x)
```

Next, use `plt.plot()` to make a **line plot**. The main arguments to `plt.plot()` are **x** data and **y** data. 

In this case, we want to plot **two** different lines (sine and cosine), so we need to supply **two** different `plt.plot()` functions. The x data are the same for both, but the y data are different (one is sine, one is cosine). 

In order to **show** the image, we need to use `plt.show()`.

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

### Changing colors, widths, and line styles'

We can add more arguments to `plt.plot()` to change the **aesthetic** of the plot. 

```python
# you should change color, linewidth, and linestyle to see what happens
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="green",  linewidth=5, linestyle="--")
plt.show()
```

### Setting limits

`plt.xlim()` and `plt.ylim()` determine the range of values that are displayed on the x and y axes. 

```python
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="red",  linewidth=2.5, linestyle="--")

# go from -10 to 10 on the x-axis
plt.xlim([-10,10])

# go from -2 to 2 on the y-axis
plt.ylim([-2,2])

plt.show()
```

### Setting ticks

Matplotlib determines the **tick marks** automatically. However, we can manually specify which tick marks we want to see. 

```python
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="red",  linewidth=2.5, linestyle="--")
plt.xlim([-5,5])
plt.ylim([-2,2])

# specify the tick marks that we want to see
plt.xticks([-np.pi,0,np.pi])
plt.yticks([-1,-0.5,0,0.5,1])

plt.show()
```

### Setting tick labels

What if we want the **tick labels** to be a different font or size?

```python
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="red",  linewidth=2.5, linestyle="--")
plt.xlim([-5,5])
plt.ylim([-1.5,1.5])

# for the xticks, we use Latex to make fancy text!
plt.xticks([-np.pi,0,np.pi],[r'$-\pi$',0,r'$\pi$'],fontsize=20) 
plt.yticks([-1,-0.5,0,0.5,1],fontsize=15)

plt.show()
```

### Moving spines

We can manually change the location and aesthetic of the **spines**. First, we create a **variable** for the axes themselves using `plt.gca()`. Then, we can modify the axes by performing functions on that variable. 

```python
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="red",  linewidth=2.5, linestyle="--")
plt.xlim([-5,5])
plt.ylim([-1.5,1.5])
plt.xticks([-np.pi,0,np.pi],[r'-$\pi$',0,r'$\pi$'],fontsize=20) 
plt.yticks([-1,-0.5,0,0.5,1],fontsize=15)

# gca stands for 'get current axis'
ax = plt.gca()  

# then, perform operations on the axes variable
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.show()
```

### Adding a legend

`plt.legend()` adds a legend to the plot. Since there are **multiple** lines on our plot, `plt.legend()` expects a **list** of labels. 

```python
plt.xlim([-5,5])
plt.ylim([-1.5,1.5])
plt.xticks([-np.pi,0,np.pi],[r'-$\pi$',0,r'$\pi$'],fontsize=20) 
plt.yticks([-1,-0.5,0,0.5,1],fontsize=15)
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, y2, color="red",  linewidth=2.5, linestyle="--")

# provide a list of labels. fontsize is optional
plt.legend(['cos','sin'],fontsize=10)
plt.show()
```

### Subplots

Commonly, we want to show **multiple** plots at once. `plt.subplot()` allows us to create **subplots**, or plots within a plot. 

`plt.subplot()` takes 3 arguments. The first 2 arguments specify the **dimensions** of our subplot. Below, we want to have a grid of 2 X 2. So, the first two arguments are (2,2). 

The third argument to `plt.subplot()` is the **specific** subplot that we are defining. For example, (2,2,1) means "in a 2 X 2 grid make the **first** plot". The order goes top to bottom, left to right. 

```python
plt.figure(figsize=(8,8))

# first plot
plt.subplot(2,2,1)
plt.plot(np.random.randn(10))

# second plot
plt.subplot(2,2,2)
plt.plot(np.random.randn(10))

# third plot
plt.subplot(2,2,3)
plt.plot(np.random.randn(10))

# fourth plot
plt.subplot(2,2,4)
plt.plot(np.random.randn(10))

plt.show()
```
